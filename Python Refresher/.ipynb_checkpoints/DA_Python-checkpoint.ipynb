```python
grade = 39
comment = "You got an"
comment1 = "You got a"

if grade >= 70:
    print(f"{comment} A")
elif grade >= 60:
    print(f"{comment1} B")
elif grade >= 50:
    print(f"{comment1} C")
elif grade >= 45:
    print(f"{comment1} D")
elif grade >= 40:
    print(f"{comment} E")
else:
    print(f"{comment1} F9")
```
