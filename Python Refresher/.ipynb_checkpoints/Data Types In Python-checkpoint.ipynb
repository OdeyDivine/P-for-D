# Introduction To Data Types in Python

##### Python has built in data types used to classify and manipulate different kinds of information. Because Python is dynamically typed, you do not need to explicitly declare a variable's type; the interpreter infers it automatically upon assignment.
##### Data types in python are already detected by the python interpreter compared to other languages like C and C++ where you have to strictly declare the data type e.g (int age = 20;)

### Core Python Data Types
##### The following are the most common data types in python in which we will be interacting majorly with
- int (Integers are whole numbers e.g 12, 19, -70)
- float (float are numbers with decimal points e.g 12.5, 17.9, 0.97)
- str (string are text data types enclosed in quotations either single or double quote e.g "Divine", 'Michael', "10", "9.4")
- bool (boleans are just frue or false data types, more like yes or no e.g True, False)

##### To find out the data type of a variable, we use the "type" function to achieve that. Find out more in the hands on lab below

```python
# run the code below to interact with data types
name = "Divine"
print(type(name))

age = 23
print(type(age))

CGPA = 4.5
print(type(age))

isNigerian = True
print(type(isNigerian))
```
